```python
activity_entries = pd.read_csv('Whitney_activity_entries.csv')
activity_types = pd.read_csv('Whitney_activity_types.csv')
hr_zones = pd.read_csv('Whitney_hr_zones.csv')
```

```python
activity_entries['start_time'].max()
```

```python
len(activity_entries)
```

```python
activity_entries.head()
```

```python
print("Each entry has its own unique ID: ",len(activity_entries) == activity_entries['logId'].nunique())
```

```python
activity_types = pd.get_dummies(data=activity_types,columns=['activity_type'])
```

```python
activity_types.reset_index(drop=True,inplace=True)
```

```python
activity_types = activity_types.drop("Unnamed: 0",axis=1)
```

```python
hr_zones = hr_zones.drop('Unnamed: 0',axis=1)
```

```python
hr_zones.head()

hr_zones[hr_zones['minutes'] > 0]
```

```python
merged_df = pd.merge(activity_entries,hr_zones, left_on="logId", right_on='log_Id',how="inner")
```
