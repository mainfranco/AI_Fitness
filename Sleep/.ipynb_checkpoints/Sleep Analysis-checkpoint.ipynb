```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge

# Dictionary of models
model_grid = {
    'GradientBoostingRegressor': GradientBoostingRegressor(),
    'SVR':SVR(),
    'Ridge':Ridge(),
    'RandomForestRegressor':RandomForestRegressor()

}
```

```python
results = {}

for model_name, model in model_grid.items():
    # Train the model
    model.fit(X_train_activitymodel, y_train_activitymodel)

    # Predict on test data
    y_pred = model.predict(X_test_activitymodel)

    # Calculate MSE and RMSE
    mse = mean_squared_error(y_test_activitymodel, y_pred)
    rmse = np.sqrt(mse)

    # Calculate R² score
    r2 = r2_score(y_test_activitymodel, y_pred)

    # Store results
    results[model_name] = {
        'Test MSE': mse,
        'Test RMSE': rmse,
        'Test R²': r2
    }

# Display results
for model_name, metrics in results.items():
    print(f"{model_name}:")
    for metric, value in metrics.items():
        print(f"   {metric}: {value}")
```

```python
gradient_boosting_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2'], 
    'subsample': [0.8, 0.9, 1.0]
}

```

```python
# Initialize the Gradient Boosting Regressor
gbr = GradientBoostingRegressor()

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=gbr, param_grid=gradient_boosting_params, cv=5, scoring='r2', verbose=2, n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(X_train_activitymodel, y_train_activitymodel)

# Get the best estimator
best_gbr = grid_search.best_estimator_

# Predict on the test set
y_pred = best_gbr.predict(X_test_activitymodel)

# Calculate metrics
mse = mean_squared_error(y_test_activitymodel, y_pred)
r2 = r2_score(y_test_activitymodel, y_pred)

# Print the results
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Test MSE: {mse}")
print(f"Test R²: {r2}")
```

```python
metrics_activitymodel = {'r2':r2, 'mse':mse}
```

```python
sleep_bedtimes['Sleep Notes'].isna().sum()
```

```python
sleep_notes
```

```python
sleep_notes_indiv = []

for i in sleep_notes:
    notes = i.split(':')
    for j in notes:
        if j not in sleep_notes_indiv:
            sleep_notes_indiv.append(j)
            
sleep_notes_indiv
```

```python
for i in sleep_notes_indiv:
    sleep_bedtimes[i] = 0

for index, row in sleep_bedtimes.iterrows():
    notes = row['Sleep Notes']
    for feature in sleep_notes_indiv:
        if feature in notes:
            sleep_bedtimes.at[index, feature] = 1

sleep_bedtimes.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import ElasticNet

# Initialize the models
models = {
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100),
    "SVR": SVR(),
    "KNN": KNeighborsRegressor(),
    "Elastic Net": ElasticNet(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42,n_estimators=100)
}

# Function to evaluate the models
def evaluate_models(models, X_train_scaled, y_train):
    for name, model in models.items():
        scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2')
        print(f"{name}: R² scores: {scores}, Mean R²: {np.mean(scores)}")

# Evaluate the models
evaluate_models(models, X_train_scaled, y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
X = np.array(sleep_bedtimes.drop(['Sleep quality'],axis=1))
y = sleep_bedtimes['Sleep quality'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train_scaled_nn = minmax_scaler.fit_transform(X_train)
X_test_scaled_nn = minmax_scaler.transform(X_test)
```

```python
num_features = len(X_train_scaled_nn[0])
```

```python
from keras.backend import clear_session
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping


best_val_loss = float('inf')
best_model = None
best_config = None


regression_loss_functions = [
    'mean_squared_error',  # Standard choice for many regression problems
    'mean_absolute_error',  # Useful when you want to minimize the effect of outliers
    'huber_loss'           # Blends MSE and MAE, offers robustness to outliers
]


regression_metrics = [
    'mean_squared_error',    # Gives a sense of the error magnitude
    'mean_absolute_error',   # Good for an average error magnitude, less sensitive to outliers
    # 'r_squared' can be added if manually defined, as it's a common evaluation metric in regression
]


for loss in regression_loss_functions:
    for metric in regression_metrics:
        clear_session()  
        model = Sequential()
        
        model.add(Dense(258, activation='relu'))
        model.add(Dense(128, activation='relu', input_shape=(num_features,)))
        model.add(Dense(64, activation='relu'))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(1, activation='linear'))


        model.compile(optimizer='adam', loss=loss, metrics=[metric])
        early_stopping = EarlyStopping(monitor='val_loss', patience=25, verbose=3, mode='min', restore_best_weights=True)


        history = model.fit(X_train_scaled_nn, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])


        current_val_loss = min(history.history['val_loss'])
        if current_val_loss < best_val_loss:
            best_val_loss = current_val_loss
            best_model = model
            best_config = {'loss': loss, 'metric': metric}

# After loop
print("Best Validation Loss:", best_val_loss)
print("Best Configuration:", best_config)

# best_model now holds the model with the best validation loss
```

```python
y_preds = best_model.predict(X_test_scaled_nn)
r2 = r2_score(y_test, y_preds)

print("R-squared Score:", r2)
```
